## CONSTRUCCION DE CARTERAS

**La primera cartera** que construiremos es la que **equitativamente distribuye** nuestro capital entre los activos de nuestras firmas.<br> Por ende, ponderaremos nuestra inversión entre ocho con **el peso de 12.5% cada una.**<br> Recordamos que la rentabilidad de esta cartera no es más que una media ponderada, o sea la rentabilidad específica de cada título por 12.5% en este caso.

En este caso obtenemos una **rentabilidad** de alrededor de **1,6%** (habiendo considerado que estaba en escala logarítmica) y una **volatilidad** de **3,5%**. <br> Nos damos cuenta de que esta no es la cartera eficiente, o sea la que subyace en la frontera eficiente de Markowitz. <br>Por consiguiente, usamos la librería de R fPortfolio para ver las diferentes opciones y construir varias combinaciones.<br>
*NOTA: Proporcionaremos también el codigo Python a seguir*

Seguramente la cartera con **mismos pesos no** es una cartera **eficiente,** el gráfico mismo nos sugiere (geométricamente) que en primer lugar esta cartera (el punto azul) **no yace en la frontera** y en segundo que el **riesgo no** está **bien asumido** porque al mismo riesgo podríamos obtener una rentabilidad mayor, por ejemplo, una mejor opción a simple vista nos sugiere que al *mismo riesgo podríamos obtener una mayor rentabilidad*, superior al 2%.<br> Por tanto, habría que cambiar el peso de los activos en nuestra cartera. La primera que está en la frontera es la de mínimo riesgo.

Una cartera con mínimo riesgo (el punto rojo en la frontera) podría ser deseable para aquellos inversores adversos al riesgo, pero preguntemos no si sería una buena opción. Por lo que vemos, considerando los activos en nuestra Cartera, respecto a la Cartera con mismos pesos el riesgo ha claramente disminuido hasta el 2,7% pero si nos fijamos mejor también se ha reducido la rentabilidad hasta 0,76%, o sea debajo el 1%.<br>
Por otro lado, si quisiéramos maximizar la ratio return/risk, o sea buscar el punto más alto en la frontera eficiente entre rentabilidad/riesgo, obtendríamos lo siguiente:


Cabe de recordar que en el caso de la Cartera de Markowitz este punto es el mismo que la ratio de Sharpe. En este caso obtenemos una rentabilidad del 3,07% pero también un riesgo alto del 4,8%. En fin, vemos el caso en el que forzamos nuestra cartera a tener una rentabilidad esperada del 2,2%, escogemos el 2,2% porque nos resulta una combinación aceptable entre la aversión completa al riesgo y el punto tangente. En el siguiente gráfico vemos los valores obtenidos:

En este caso la rentabilidad la hemos claramente establecido, y el riesgo resulta ser más bajo: 3,6%.


#### En Python:

```python
from pypfopt.efficient_frontier import EfficientFrontier
from pypfopt import risk_models, expected_returns
```

```python
# Rentabilidad mensual
rentabilidad = portfolio.iloc[:,:-1].pct_change(periods=20).dropna().mean()
rentabilidad 
```

```python
w1 = np.repeat(0.125, 8)     # Pesos de los activo: Suponemos una inversiòn por igual de ellos inicial.
cov_matrix = portfolio.iloc[:,:-1].pct_change(periods=20).dropna().cov()*20 # Matriz de covarianza
cov_matrix
```

```python
# Varianza de la cartera 
cartera_var = np.dot(w1.T, np.dot(cov_matrix, w1))
cartera_var
```

```python
# Calculamos la volatilidad osea la Desviaciòn
volatilidad = np.sqrt(cartera_var)
volatilidad
```

```python
# Calculamos la rentabilidad anual
rentabilidad_anual = np.sum(rentabilidad * w1)*20
rentabilidad_anual
```

```python
# Resumiendo 
p_var = str(round(cartera_var, 2)*100)+'%'
p_vol = str(round(volatilidad, 2)*100)+'%'
p_ret = str(round(rentabilidad_anual, 2)*100)+'%'

print('Rentabilidad esperada: ', p_ret)
print('Volatilidad/Riesgo: ', p_vol)
print('Varianza: ', p_var)
```

**Portfolio Optimization Package**

Cartera agresiva: Usamos la ratio de return $\to$ risk Sharpe

```python
mu = expected_returns.mean_historical_return(portfolio.iloc[:,:-1])
S = risk_models.sample_cov(portfolio.iloc[:,:-1])
```

```python
# Optimizar segun la ratio de Sharpe
ef = EfficientFrontier(mu, S)
pesos_sharpe = ef.max_sharpe()
peso = ef.clean_weights()
print(peso)
```

max_sharpe $\to$ Maximiza la relación de Sharpe. El resultado también se conoce como la cartera de tangencia, ya que es la cartera para la cual la línea del mercado de capitales es tangente a la frontera eficiente.

Como veemos esta cartera es agresiva o sea max. el retorno de la inversiòn. Por tanto, veemos una volatilidad discretamente alta.<br> Por otro lado, veemos que no deberìamos invertir en algunos activos si quicieramos max. la rentabilidad.

```python
ef.portfolio_performance(verbose=True)
```

#### Veamos en concreto las acciones que deberìamos comprar si quicieramos gestionar una cartera con una retanbilidad considerable.
De hecho esto es interesante pero no útil en sí mismo. Sin embargo, PyPortfolioOpt proporciona un método que le permite convertir los pesos continuos anteriores en una asignación real que puede comprar. <br>Simplemente debemos ingresar los precios más recientes y el tamaño de cartera deseado ($ 15,000 en este ejemplo):

```python
from pypfopt.discrete_allocation import DiscreteAllocation, get_latest_prices
```

```python
ultimo_precio = get_latest_prices(portfolio.iloc[:,:-1])
ultimo_precio # ultimo precio de nuestros activos
```

Recordemos los pesos

```python
pes = peso
allocation = DiscreteAllocation(pes, ultimo_precio, total_portfolio_value=15000) # Inversiòn de $15k
activos, leftover = allocation.lp_portfolio()
print('Cartera agresiva, Cantidad de activos que deberìa comprar: ',activos)
print('Lo que me queda: ${:.2f}'.format( leftover))
```

## Aversiòn al riesgo 

Imaginemos ahora que somos "cuidadosos" o simplemente somos aversos al riesgo. Bien esto es lo que deberìamos hacer:

```python
mu2 = expected_returns.mean_historical_return(portfolio.iloc[:,:-1])
S2 = risk_models.sample_cov(portfolio.iloc[:,:-1])
```

```python
ef_min = EfficientFrontier(mu2, S2)
pesos_min = ef_min.min_volatility()
peso2 = ef_min.clean_weights()
print(peso2)
```

```python
ef_min.portfolio_performance(verbose=True)
```

Como podemos ver el riesgo ha bajado considerablemente. Hay pero que constatar que la rentabilidad es mucho mas baja respecto al riesgo.<br> Y en fin la ratio de Sharpe no es ni siquiera de la unidad, por lo que la cartera "agresiva" resulta ser mejor a paridad de riesgo.

#### Veamos de todas manera como deberìamos escoger los activos si estuvieramos interesados en esta cartera

```python
allocation_min = DiscreteAllocation(pesos_min, ultimo_precio, total_portfolio_value=15000) # Inversiòn de $15k
activos_min, leftover_min = allocation_min.lp_portfolio()
print('Cartera agresiva, Cantidad de activos que deberìa comprar: ',activos_min)
print('Lo que me queda: ${:.2f}'.format( leftover_min))
```

Tal como lo esperabamos. Ahora deberiamos bajar la acciones de **Planet Fitness.** <br>Por tanto, con una cartera de **minimo riesgo** deberiamos: **comprar 13 de TXN, 21 de TYL, 4 de INTC y 71 de ORCL.**
